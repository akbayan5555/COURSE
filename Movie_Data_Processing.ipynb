```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import pandas as pd

# Загрузим данные
df = pd.read_csv('movie_metadata.csv')

# Отбор нужных столбцов
df = df[['movie_title', 'imdb_score', 'genres']]
df['high_rating'] = (df['imdb_score'] > 7).astype(int)

# Преобразование жанров в бинарные признаки
df = pd.concat([df, df['genres'].str.get_dummies(sep='|')], axis=1)
df.drop(['movie_title', 'genres', 'imdb_score'], axis=1, inplace=True)

# Разделим данные на признаки и целевую переменную
X = df.drop('high_rating', axis=1)
y = df['high_rating']

# Разделим на обучающую и тестовую выборки
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Инициализация модели Random Forest
rf = RandomForestClassifier(random_state=42)

# Обучение модели
rf.fit(X_train, y_train)

# Прогнозирование на тестовых данных
y_pred = rf.predict(X_test)

# Оценка точности модели
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Визуализация результатов (если нужно)
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier

# Визуализация важности признаков
feature_importances = rf.feature_importances_
features = X.columns

plt.figure(figsize=(10, 6))
plt.barh(features, feature_importances)
plt.xlabel('Feature Importance')
plt.title('Random Forest Feature Importance')
plt.show()
```
